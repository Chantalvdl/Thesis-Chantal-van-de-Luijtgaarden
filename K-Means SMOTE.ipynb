```python
# Split data in X and y
def X_Y(train, labelss):
    X_train = train.drop(labels=labelss, axis=1)
    X_train = X_train.to_numpy()
    y_train = train["label"].to_numpy()
    return X_train, y_train
```

```python
# All functions
# Get oversampled data back for both X and y
def balancing(data, n, thres, kn, sas):
    labels = ["label"]
    X, y = X_Y(data, labels)
    sm = KMeansSMOTE(cluster_balance_threshold=thres, k_neighbors=kn, kmeans_estimator=n, sampling_strategy=sas, n_jobs=-1)
    try: 
        X_res, y_res = sm.fit_resample(X, y)
    except RuntimeError:
        X_res, y_res = X, y
        print("nope")
    print(np.count_nonzero(y_res == 0))
    print(np.count_nonzero(y_res == 1))
    print(sm.get_params())
    return X_res, y_res

# Reshape the data in new format
def update_frame(data, X_res, y_res):
    new = np.reshape(X_res, (len(X_res), len(X_res[0]))).T
    finaldata = pd.DataFrame(index = None)
    columns = data.columns
    for column, col in zip(new, columns):
        finaldata[col] = column 
    finaldata["label"] = y_res
    return finaldata

# Create final clusters
def clusters(data, n, inits, ninit, ite):
    good = True
    kmeans = KMeans(init=inits, n_clusters=n, n_init=ninit, max_iter=ite)
    labels = ["label"]
    X, y = X_Y(data, labels)
    result = kmeans.fit_predict(X,y)
    data["cluster"] = result
    for i in range(0,n):
        filtr = data[(data.cluster == i)]
        print("Negative:",str(i),(len(filtr) - sum(filtr.label)))
        print("Positive:", " ", sum(filtr.label))
        print()   
        if good != False:
            if (abs((len(filtr) - sum(filtr.label)) - sum(filtr.label)) / len(filtr)) > 0.25:
                good = False
    return data, good

# Dropping columns from data
def adjusting_data(data, columns):
    data = data[columns]
    return data
```

```python
# Adjust the threshold for Good!
def clusters(data, n, inits, ninit, ite):
    good = True
    kmeans = KMeans(init=inits, n_clusters=n, n_init=ninit, max_iter=ite)
    labels = ["label"]
    X, y = X_Y(data, labels)
    result = kmeans.fit_predict(X,y)
    data["cluster"] = result
    for i in range(0,n):
        filtr = data[(data.cluster == i)]
        print("Negative:",str(i),(len(filtr) - sum(filtr.label)))
        print("Positive:", " ", sum(filtr.label))
        print()   
        if good != False:
            if (abs((len(filtr) - sum(filtr.label)) - sum(filtr.label)) / len(filtr)) > 0.35:
                good = False
    return data, good
```
