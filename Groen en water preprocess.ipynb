```python
# For all datasets sum all surfaces of different green, water and road components
datasets = [groen18, groen19, water18, water19, straat18, straat19, groen17, water17, straat17]
sets = []
indexes = []
i = 0
for dataset in datasets:
    print(dataset)
    for buurt in dataset.buurt_code.unique():
        if i % 10000 == 0:
            print(i)
        i = i + 1
        filt = dataset[(dataset.buurt_code == buurt)]
        if len(filt) > 1:
            column = filt.columns[1]
            total = sum(filt[column])
            index = filt.iloc[0].name
            dataset[column].iloc[index] = total
            ind = filt.iloc[1:].index
            for j in ind:
                indexes.append(j)
    dataset = dataset[~(dataset.index.isin(indexes))]
    indexes = []
    print(len(dataset))
    sets.append(dataset)

groen18 = sets[0] 
groen19 = sets[1] 
water18 = sets[2] 
water19 = sets[3] 
straat18 = sets[4] 
straat19 = sets[5]
groen17 = sets[6] 
water17 = sets[7] 
straat17 = sets[8]

groen18.to_csv("2018\\Groen_2018_filtered.csv", index=False)
groen19.to_csv("2019\\Groen_2019_filtered.csv", index=False)
water18.to_csv("2018\\Water_2018_filtered.csv", index=False)
water19.to_csv("2019\\Water_2019_filtered.csv", index=False)
straat18.to_csv("2018\\Weg_2018_filtered.csv", index=False)
straat19.to_csv("2019\\Weg_2019_filtered.csv", index=False)
groen17.to_csv("2017\\Groen_2017_filtered.csv", index=False)
water17.to_csv("2017\\Water_2017_filtered.csv", index=False)
straat17.to_csv("2017\\Weg_2017_filtered.csv", index=False)
```
