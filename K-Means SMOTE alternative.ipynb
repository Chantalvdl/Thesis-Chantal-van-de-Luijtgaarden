```python
import pandas as pd
import numpy as np
import math
from imblearn.over_sampling import KMeansSMOTE
from imblearn.over_sampling import SMOTE
from sklearn.cluster import KMeans
```

```python
# Load dataset
Data_2017 = pd.read_csv("..\\Data\\old\\17_final.csv", delimiter=',')
Data_2018 = pd.read_csv("..\\Data\\old\\18_final.csv", delimiter=',')
Data_2019 = pd.read_csv("..\\Data\\old\\19_final.csv", delimiter=',')
datas = [Data_2017 , Data_2018, Data_2019]
```

# Check imbalance


```python
# Observe class imbalance
print(sum(Data_2017.label == 1))
print(sum(Data_2017.label == 0))
print(sum(Data_2017.label == 1)/len(Data_2017))
print(sum(Data_2017.label == 0)/len(Data_2017))

# Observe class imbalance
print(sum(Data_2018.label == 1))
print(sum(Data_2018.label == 0))
print(sum(Data_2018.label == 1)/len(Data_2018))
print(sum(Data_2018.label == 0)/len(Data_2018))

# Observe class imbalance
print(sum(Data_2019.label == 1))
print(sum(Data_2019.label == 0))
print(sum(Data_2019.label == 1)/len(Data_2019))
print(sum(Data_2019.label == 0)/len(Data_2019))
```

```python
# Split data in X and y
def X_Y(train, labelss):
    X_train = train.drop(labels=labelss, axis=1)
    X_train = X_train.to_numpy()
    y_train = train["label"].to_numpy()
    return X_train, y_train
```

```python
# Try KMeans on the different datasets
n = 5
results = []
kmeans = KMeans(init="k-means++", n_clusters=n, n_init=30, max_iter=800)
labels = ["c28992r100", "AANTAL_HH","GEM_HH_GR","P_HUURWON","P_KOOPWON", "G_ELEK_WON", 'WON_NBEW',
          "G_GAS_WON", "WOZWONING", "woon","winkel","gezondheid","kantoor","bijeenkomst",
          "onderwijs","sport","industrie","logies","min_construction_year","max_construction_year","mean_construction_year",
          "P_HINK_HH","P_LINK_HH", "UITKMINAOW", "sur_green","sur_water","sur_road","sur_other",
         "Fundering",'INW_014', 'INW_1524', 'INW_2544', 'INW_4564', 'INW_65PL']

for data in datas:
    X, y = X_Y(data, labels)
    result = kmeans.fit_predict(X,y)
    results.append(result)
    data["cluster"] = result
    print(result)
    for i in range(0,n):
        filtr = data[(data.cluster == i)]
        print("Cluster " + str(i))
        print("Negative:",(len(filtr) - sum(filtr.label)))
        print("Positive:", " ", sum(filtr.label))
        print("--------")
```

```python
qgis_train = pd.DataFrame(Data_2017["c28992r100"],columns=["c28992r100"])
qgis_test = pd.DataFrame(Data_2018["c28992r100"],columns=["c28992r100"])
qgis_val = pd.DataFrame(Data_2019["c28992r100"],columns=["c28992r100"])
qgis_train["cluster"] = results[0]
qgis_test["cluster"] = results[1]
qgis_val["cluster"] = results[2]

qgis_train.to_csv("qgistrainno.csv", index=False)
qgis_test.to_csv("qgistestno.csv", index=False)
qgis_val.to_csv("qgisvalno.csv", index=False)
```

```python
# Do SMOTE separately
samples = []
labels = ["c28992r100","label"]

data = datas[0]
X, y = X_Y(data,labels)
smote = SMOTE(k_neighbors=10, sampling_strategy=1.)
X_res, y_res = smote.fit_resample(X, y)
samples.append([X_res, y_res])
print(np.count_nonzero(y_res == 0))
print(np.count_nonzero(y_res == 1))
print(smote.get_params())

data = datas[1]
X, y = X_Y(data,labels)
smote = SMOTE(k_neighbors=10, sampling_strategy=1.0)
X_res, y_res = smote.fit_resample(X, y)
samples.append([X_res, y_res])
print(np.count_nonzero(y_res == 0))
print(np.count_nonzero(y_res == 1))
print(smote.get_params())

data = datas[2]
X, y = X_Y(data,labels)
smote = SMOTE(k_neighbors=10, sampling_strategy=1.0)
X_res, y_res = smote.fit_resample(X, y)
samples.append([X_res, y_res])
print(np.count_nonzero(y_res == 0))
print(np.count_nonzero(y_res == 1))
print(smote.get_params())

# Reshape results
feature_columns = []
for sa in samples:
    new = np.reshape(sa[0], (len(sa[0]), len(sa[0][0]))).T
    print(new)
    print(len(new))
    feature_columns.append([new,sa[1]])
    print("---------")
```
