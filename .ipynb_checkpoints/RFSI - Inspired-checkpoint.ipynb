# Get Nearest Neighbors

```python
# Splits data in Train and Testset based
def train_test_splitter(dataset, predictor, nullvalue = -99997):
    # Get part of dataset without null values
    train = dataset[(dataset[predictor] != nullvalue)].reset_index()
    # Get part of dataset with null values to predict
    test = dataset[(dataset[predictor] == nullvalue)].reset_index()
    return train, test

# Appending 4 nearest neighbor out, based on the training data
def nearest_neighbors_cov(dataset, neighborset, n = 4):
    # Define empty lists
    neighbors = []
    items = []
    # Iterate over the dataset
    for index, grid in dataset.iterrows():
        items = []
        its = []
        i = 1
        # Indicator for where the algo is
        if index % 10000 == 0:
            print("index:" + str(index))
        # If the list of items is not filled with at least 4 items, keep expanding the search
        while len(items) < n:
            # filter on the right columnnumbers
            cfilter = neighborset[((neighborset["row"]<(int(grid["row"])+i)) & (neighborset["row"]> (int(grid["row"])-i)))]
            # filter on the right rownumbers
            rfilter = neighborset[((neighborset["column"]<(int(grid["column"])+ i)) & (neighborset["column"]>(int(grid["column"])-i)))]
            # Find the common gridcells
            own = pd.merge(cfilter, rfilter['c28992r100'], how='inner', on=['c28992r100'])
            # Remove the cell that we are finding the neighbors for
            n_grids = own[(own.c28992r100 != grid["c28992r100"])]
            # Remove the neighbors that we have found in previous searches to avoid duplicates
            n_grids = n_grids[~(n_grids.c28992r100.isin(items))]
            # if there are neighbors found then append them to items
            if len(n_grids) != 0:
                for index, item in n_grids.iterrows():
                    its = []
                    # append every needed attribute
                    for it in item:
                        its.append(it)
                    items.append(its)
            i = i + 1
        # If items contain exactly the number of neighbors needed
        if len(items) == n:
            neighbors.append(items)
        # If items contain more than the number of neighbors needed, only get the top 4, 
        # ensuring earlier found neighbors get priority
        elif len(items) > n:
            neighbors.append(items[:4]) 
    return neighbors

# Reshape the data in a dataframe including the neighbors
def reshape_data(cols, neighbordata, dataset, columns, pred, year):
    neighborhoods = []
    cs = []
    for ind, data in dataset.iterrows():
        neighborhood = []
        # Append every feature separate again
        for d in data:
            neighborhood.append(d)
        for neigh in neighbordata[ind]:
            # Append all the features of the original and its neighbors in one list, expect for the first two
            for n in neigh[2:]:
                neighborhood.append(n)
        neighborhoods.append(neighborhood)
    print((len(neighborhoods),len(neighborhoods[0])))
    # Reshape the arrays in dataframe shape
    neighborhoods = np.reshape(neighborhoods, ((len(neighborhoods),len(neighborhoods[0])))).T
    i = 0 
    # Create all column and columnnames for neighbordata
    for c in dataset.columns:
        cs.append(c)
    for i in range(1,5):
        for c in cols:
            column_name = c + "_" + str(i)
            cs.append(column_name)
    print(cs)
    # Create dataframe
    complete_data = pd.DataFrame(data, columns=[cs])
    for c, clms in zip(range(0,len(cs)), cs):
        complete_data[c] = neighborhoods[c]
    # Create csv with correct namee
    name = pred + "_" + str(year) + ".csv"
    complete_data.to_csv(name, index=False)
    return complete_data
    
    
def main(dataset, pred, year, n):
    cols = dataset.columns[1:]
    train, test = train_test_splitter(dataset, pred)
    # Neighbors for train and test separately, since neighbors can not come from test set for both train and test
    train_nc = nearest_neighbors_cov(train, train, n)
    test_nc = nearest_neighbors_cov(test, train, n)
    traindata = reshape_data(cols, train_nc, train, dataset, pred, year)
    # Make new year string to avoid csv name collision
    year = str(year) + "test"
    testdata = reshape_data(cols, test_nc, test, dataset, pred, year)
    
```

```python
# Calculate Distance
def distance(row, i):
    rn = "row_" + str(i)
    cn = "column_" + str(i)
    ro = row["row"]
    co = row["column"]
    r = row[rn]
    c = row[cn]
    dis = math.sqrt((r-ro)**2+(c-co)**2)
    return dis

# Get the distance in correct format for the entire dataframe
def loop(data):
    for index, row in data.iterrows():
        if index%80000==0:
            print(index)
        dis = distance(row, 1)
        dis_1.append(round(dis,2))
        dis = distance(row, 2)
        dis_2.append(round(dis,2))
        dis = distance(row, 3)
        dis_3.append(round(dis,2))
        dis = distance(row, 4)
        dis_4.append(round(dis,2))
    return dis_1, dis_2, dis_3, dis_4

```

```python
def train_val(dataset):
    if len(dataset)> 300000:
        train = dataset.sample(n=225000)
    else:
        train = dataset.sample(frac=0.75)
    test = dataset[~(dataset.c28992r100.isin(train["c28992r100"]))]
    val = test.sample(frac=0.9) 
    val2 = test[~(test.c28992r100.isin(val["c28992r100"]))]
    return train, val, val2

# Creating Models with possible parameters
def objective(space):
    classifier = RandomForestRegressor(n_estimators = space['n_estimators'],
                            max_depth = int(space['max_depth']),
                            min_samples_leaf = int(space['min_samples_leaf']),
                            min_samples_split = space['min_samples_split'],
                            n_jobs = -1)
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_train)
    accuracy_score = mean_absolute_error(y_train, pred)
    
    predictions = classifier.predict(X_val)
    accuracy_score_metric = mean_absolute_error(y_val, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)
    
    predictions = classifier.predict(X_val2)
    accuracy_score_metric = mean_absolute_error(y_val2, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)

    return{'loss':accuracy_score, 'status': STATUS_OK }

def X_Y(train, val, val2):
    X_train = train.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_train = X_train.to_numpy()
    y_train = train[[label]].to_numpy()
    
    X_val = val.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val = X_val.to_numpy()
    y_val = val[[label]].to_numpy()
    
    X_val2 = val2.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val2 = X_val2.to_numpy()
    y_val2 = val2[[label]].to_numpy()
    return X_train, y_train, X_val ,y_val, X_val2 ,y_val2
```

```python
variablenam = []
for i in variablenam_17:
    variablenam.append(i)
for i in variablenam_18:
    variablenam.append(i)
for i in variablenam_19:
    variablenam.append(i)
    
variablelists = []
for i in variablelists_17:
    variablelists.append(i)
for i in variablelists_18:
    variablelists.append(i)
for i in variablelists_19:
    variablelists.append(i)
```

```python
rfs = []
ne = []
for x in range(10,100,5):
    ne.append(x)
md = []
for x in range(2,20,2):
    md.append(x)
mss = []
for x in range(2,30,3):
    mss.append(x)

for name, var in zip(variablenam,variablelists):
    train, val, val2 = train_val(var)
    label = (name.split('_1'))[0]
    X_train, y_train, X_val ,y_val, X_val2 ,y_val2 = X_Y(train, val, val2)
    space = {'n_estimators':hp.choice('n_estimators',ne),
               'max_depth':hp.choice('max_depth',md),
               'min_samples_leaf':hp.quniform('min_samples_leaf',1,10,1),
               'min_samples_split':hp.choice('min_samples_split',mss)}

    trials = Trials()
    best = fmin(fn=objective,
                space=space,
                algo=tpe.suggest,
                max_evals=25,
                trials=trials,
               verbose = 1)

    print("Best: ", best)


    # Fitting XGBoost to the Training set
    classifier = RandomForestRegressor(n_estimators = ne[best['n_estimators']],
                                max_depth = md[best['max_depth']],
                                min_samples_leaf = int(best['min_samples_leaf']),
                                min_samples_split= mss[best['min_samples_split']],
                               n_jobs = -1,)
    print(name)
    print(classifier)

    rfs.append([classifier, train, val, val2])
```

```python
from hyperopt import *
from sklearn.metrics import mean_absolute_error
import pyspark 
import mlflow
```

```python
def train_val(dataset):
    if len(dataset)> 300000:
        train = dataset.sample(n=225000)
    else:
        train = dataset.sample(frac=0.75)
    test = dataset[~(dataset.c28992r100.isin(train["c28992r100"]))]
    val = test.sample(frac=0.9) 
    val2 = test[~(test.c28992r100.isin(val["c28992r100"]))]
    return train, val, val2

# Creating Models with possible parameters
def objective(space):
    classifier = RandomForestRegressor(n_estimators = space['n_estimators'],
                            max_depth = int(space['max_depth']),
                            min_samples_leaf = int(space['min_samples_leaf']),
                            min_samples_split = space['min_samples_split'],
                            n_jobs = -1)
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_train)
    accuracy_score = mean_absolute_error(y_train, pred)
    
    predictions = classifier.predict(X_val)
    accuracy_score_metric = mean_absolute_error(y_val, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)
    
    predictions = classifier.predict(X_val2)
    accuracy_score_metric = mean_absolute_error(y_val2, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)

    return{'loss':accuracy_score, 'status': STATUS_OK }

def X_Y(train, val, val2):
    X_train = train.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_train = X_train.to_numpy()
    y_train = train[[label]].to_numpy()
    
    X_val = val.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val = X_val.to_numpy()
    y_val = val[[label]].to_numpy()
    
    X_val2 = val2.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val2 = X_val2.to_numpy()
    y_val2 = val2[[label]].to_numpy()
    return X_train, y_train, X_val ,y_val, X_val2 ,y_val2
```

```python
rfs = []
ne = []
for x in range(10,100,5):
    ne.append(x)
md = []
for x in range(2,20,2):
    md.append(x)
mss = []
for x in range(2,30,3):
    mss.append(x)

for name, var in zip(variablenam,variablelists):
    train, val, val2 = train_val(var)
    label = (name.split('_1'))[0]
    X_train, y_train, X_val ,y_val, X_val2 ,y_val2 = X_Y(train, val, val2)
    space = {'n_estimators':hp.choice('n_estimators',ne),
               'max_depth':hp.choice('max_depth',md),
               'min_samples_leaf':hp.quniform('min_samples_leaf',1,10,1),
               'min_samples_split':hp.choice('min_samples_split',mss)}

    trials = Trials()
    best = fmin(fn=objective,
                space=space,
                algo=tpe.suggest,
                max_evals=25,
                trials=trials,
               verbose = 1)

    print("Best: ", best)


    # Fitting XGBoost to the Training set
    classifier = RandomForestRegressor(n_estimators = ne[best['n_estimators']],
                                max_depth =  md[best['max_depth']],
                                min_samples_leaf = int(best['min_samples_leaf']),
                                min_samples_split= mss[best['min_samples_split']],
                               n_jobs = -1,)
    print(name)
    print(classifier)

    rfs.append([classifier, train, val, val2])
```

```python
def train_val(dataset):
    if len(dataset)> 300000:
        train = dataset.sample(n=225000)
    else:
        train = dataset.sample(frac=0.75)
    test = dataset[~(dataset.c28992r100.isin(train["c28992r100"]))]
    val = test.sample(frac=0.9) 
    val2 = test[~(test.c28992r100.isin(val["c28992r100"]))]
    return train, val, val2

# Creating Models with possible parameters
def objective(space):
    classifier = RandomForestRegressor(n_estimators = space['n_estimators'],
                            max_depth = int(space['max_depth']),
                            min_samples_leaf = int(space['min_samples_leaf']),
                            min_samples_split = space['min_samples_split'],
                            n_jobs = -1)
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_train)
    accuracy_score = mean_absolute_error(y_train, pred)
    
    predictions = classifier.predict(X_val)
    accuracy_score_metric = mean_absolute_error(y_val, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)
    
    predictions = classifier.predict(X_val2)
    accuracy_score_metric = mean_absolute_error(y_val2, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)

    return{'loss':accuracy_score, 'status': STATUS_OK }

def X_Y(train, val, val2):
    X_train = train.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_train = X_train.to_numpy()
    y_train = train[[label]].to_numpy()
    
    X_val = val.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val = X_val.to_numpy()
    y_val = val[[label]].to_numpy()
    
    X_val2 = val2.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val2 = X_val2.to_numpy()
    y_val2 = val2[[label]].to_numpy()
    return X_train, y_train, X_val ,y_val, X_val2 ,y_val2

rfs = []
ne = []
for x in range(10,100,5):
    ne.append(x)
md = []
for x in range(2,20,2):
    md.append(x)
mss = []
for x in range(2,30,3):
    mss.append(x)

for name, var in zip(variablenames,variablelists):
    train, val, val2 = train_val(var)
    label = (name.split('_1'))[0]
    X_train, y_train, X_val ,y_val, X_val2 ,y_val2 = X_Y(train, val, val2)
    space = {'n_estimators':hp.choice('n_estimators',ne),
               'max_depth':hp.choice('max_depth',md),
               'min_samples_leaf':hp.quniform('min_samples_leaf',1,10,1),
               'min_samples_split':hp.choice('min_samples_split',mss)}

    trials = Trials()
    best = fmin(fn=objective,
                space=space,
                algo=tpe.suggest,
                max_evals=25,
                trials=trials,
               verbose = 1)

    print("Best: ", best)


    # Fitting XGBoost to the Training set
    classifier = RandomForestRegressor(n_estimators = ne[best['n_estimators']],
                                max_depth =  md[best['max_depth']],
                                min_samples_leaf = int(best['min_samples_leaf']),
                                min_samples_split= mss[best['min_samples_split']],
                               n_jobs = -1,)
    print(name)
    print(classifier)
    rfs.append([classifier, train, val, val2])
```

```python
def train_val(dataset):
    if len(dataset)> 300000:
        train = dataset.sample(n=225000)
    else:
        train = dataset.sample(frac=0.75)
    test = dataset[~(dataset.c28992r100.isin(train["c28992r100"]))]
    val = test.sample(frac=0.9) 
    val2 = test[~(test.c28992r100.isin(val["c28992r100"]))]
    return train, val, val2

# Creating Models with possible parameters
def objective(space):
    classifier = RandomForestRegressor(n_estimators = space['n_estimators'],
                            max_depth = int(space['max_depth']),
                            min_samples_leaf = int(space['min_samples_leaf']),
                            min_samples_split = space['min_samples_split'],
                            n_jobs = -1)
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_train)
    accuracy_score = mean_absolute_error(y_train, pred)
    
    predictions = classifier.predict(X_val)
    accuracy_score_metric = mean_absolute_error(y_val, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)
    
    predictions = classifier.predict(X_val2)
    accuracy_score_metric = mean_absolute_error(y_val2, predictions)
    print("Accuracy_score_metric:", accuracy_score_metric)

    return{'loss':accuracy_score, 'status': STATUS_OK }

def X_Y(train, val, val2):
    X_train = train.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_train = X_train.to_numpy()
    y_train = train[[label]].to_numpy()
    
    X_val = val.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val = X_val.to_numpy()
    y_val = val[[label]].to_numpy()
    
    X_val2 = val2.drop(labels=[label,"c28992r100","row","column", "row_1","column_1", "row_2","column_2", "row_3",
                                 "column_3", "row_4","column_4","index"], axis=1)
    X_val2 = X_val2.to_numpy()
    y_val2 = val2[[label]].to_numpy()
    return X_train, y_train, X_val ,y_val, X_val2 ,y_val2

rfs = []
ne = []
for x in range(10,100,5):
    ne.append(x)
md = []
for x in range(2,20,2):
    md.append(x)
mss = []
for x in range(2,30,3):
    mss.append(x)

for name, var in zip(variablenames,variablelists):
    train, val, val2 = train_val(var)
    label = (name.split('_1'))[0]
    X_train, y_train, X_val ,y_val, X_val2 ,y_val2 = X_Y(train, val, val2)
    space = {'n_estimators':hp.choice('n_estimators',ne),
               'max_depth':hp.choice('max_depth',md),
               'min_samples_leaf':hp.quniform('min_samples_leaf',1,10,1),
               'min_samples_split':hp.choice('min_samples_split',mss)}

    trials = Trials()
    best = fmin(fn=objective,
                space=space,
                algo=tpe.suggest,
                max_evals=25,
                trials=trials,
               verbose = 1)

    print("Best: ", best)


    # Fitting XGBoost to the Training set
    classifier = RandomForestRegressor(n_estimators = ne[best['n_estimators']],
                                max_depth =  md[best['max_depth']],
                                min_samples_leaf = int(best['min_samples_leaf']),
                                min_samples_split= mss[best['min_samples_split']],
                               n_jobs = -1,)
    print(name)
    print(classifier)

    rfs.append([classifier, train, val, val2])
```
